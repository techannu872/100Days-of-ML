## **Considering the Whole Container(for all pages)**

```python
final_df=pd.DataFrame()
for j in range(0):

  webpage=requests.get('https://internshala.com/internships?page={}'.format(j)).text

  soup=BeautifulSoup(webpage,'lxml')

  company=soup.find_all('div',class_='internship_meta')

  job_role=[]
  name_of_company=[]
  location=[]
  duration=[]
  stipend=[]

  for i in company:
    try:
      job_role.append(i.find('h3').text.strip())
    except:
      job_role.append(np.nan)

    try:
      name_of_company.append(i.find('a' , class_='link_display_like_text view_detail_button').text.strip())
    except:
      name_of_company.append(np.nan)

    try:
      location.append(i.find('a', class_='location_link view_detail_button').text.strip())
    except:
      location.append(np.nan)

    try:
      duration.append(i.find('div', class_='item_body').text.strip())
    except:
      duration.append(np.nan)

    try:
      stipend.append(i.find('span', class_='stipend').text.strip())
    except:
      stipend.append(np.nan)

  df=pd.DataFrame({'Job_Role':job_role, 'Company_Name':name_of_company, 'Location':location, 'Duration':duration, 'Stipend':stipend})

  final_dataset=final_df.append(df,ignore_index=True)
















```
