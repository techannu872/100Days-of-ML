## **Opening CSV(Comma Separated Values) files by converting feature's paramete using convertor**

```python
pd.read_csv('/content/IPL Matches 2008-2020.csv')
```

```python
def rename(name): #Here we defined the function to rename the parameter
    if name == "Chennai Super Kings":
        return "CSK Thala for a reason"
    else:
        return name
rename("Chennai Super Kings")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team2':rename}) #Here we used convertor par and used the func name we defined above
```
